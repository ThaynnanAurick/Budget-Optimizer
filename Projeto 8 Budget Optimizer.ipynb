````python

# # Extract time features from 'Start Ts'
visits['session_month'] = visits['start_ts'].astype('datetime64[M]')
visits['session_date'] = visits['start_ts'].dt.date

# Merge the first session information back to the visits dataset
vis = pd.merge(visits, first_visits, on='uid')

# Calculate age (time difference) in days and months
vis['age_days'] = vis['session_date'] - vis['first_session_dt']
vis['age_months'] = ((vis['session_month'] - vis['first_session_month']) / np.timedelta64(1, 'M')).round().astype('int')
vis['first_session_month'] = vis['first_session_month'].dt.strftime('%Y-%m')

# Create a cohort pivot table
cohorts = vis.pivot_table(
    index='first_session_month',
    columns='age_months',
    values='uid',
    aggfunc='nunique'
)

# Fill missing values in the cohort table
cohorts = cohorts.fillna(0)

# Calculate retention rates
retention = pd.DataFrame()
for col in cohorts.columns:
    retention[col] = cohorts[col] / cohorts[0]

# Fill missing values with an empty string for display purposes
retention = retention.fillna('')

# Convert retention rates to percentage
retention_percentage = (retention * 100).round(0)

# Create a mask for 0% values (True where values are 0)
mask = retention_percentage == 0

# Plot the heatmap
plt.figure(figsize=(16, 10))
ax = sns.heatmap(
    retention_percentage,
    annot=True,  # Add annotations
    fmt=".0f",  # Format annotations as percentages
    cmap="YlGnBu",  # Use a green-blue color map
    mask=mask,  # Apply mask to hide 0% values and color those cells white
    cbar_kws={'label': 'Retention Rate (%)'},  # Add color bar label
    linewidths=0.5,  # Add gridlines
    linecolor="white"  # Set gridlines to white for better contrast
)

# Add labels and title
ax.set_title("User Retention by Cohort", fontsize=18)
ax.set_xlabel("Months Since First Login", fontsize=14)
ax.set_ylabel("Cohort Month", fontsize=14)

# Show the plot
plt.tight_layout()
````

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Read in the data (already in your case)
# visits = pd.read_csv('/datasets/visits_log_us.csv')
# orders = pd.read_csv('/datasets/orders_log_us.csv')
# costs = pd.read_csv('/datasets/costs_us.csv')

# 2. Convert the relevant columns to datetime format
orders['buy_ts'] = pd.to_datetime(orders['buy_ts'])
costs['dt'] = pd.to_datetime(costs['dt'])

# 3. Create 'order_month' and 'first_order_month' columns in orders
orders['order_month'] = orders['buy_ts'].dt.to_period('M')
orders['first_order_month'] = orders.groupby('uid')['order_month'].transform('min')

# 4. Group the costs and orders data by month
costs['costs_month'] = costs['dt'].dt.to_period('M')

# 5. Calculate CAC per month by merging costs and orders
CAC_por_mes = pd.merge(
    costs.groupby('costs_month').agg({'costs': 'sum'}),
    orders.groupby('first_order_month').agg({'uid': 'nunique'}),  # Number of unique users
    left_index=True,
    right_index=True
)

# 6. Calculate the Cost of Acquisition (CAC)
CAC_por_mes['cac'] = CAC_por_mes['costs'] / CAC_por_mes['uid']

# 7. Plot the CAC over time
CAC_por_mes['cac'].plot(figsize=(14,7), title='CAC Médio por mês', xlabel='Mês', ylabel='CAC (c.u.)')
plt.show()

# 8. Create cohorts based on the first_order_month and calculate revenue and number of buyers
cohorts = orders.groupby(['first_order_month', 'order_month']).agg(
    revenue=('revenue', 'sum'),
    n_buyers=('uid', 'nunique')  # Unique buyers
).reset_index()

# 9. Calculate the cohort lifetime in months
cohorts['cohort_lifetime'] = (
    (cohorts['order_month'].dt.year - cohorts['first_order_month'].dt.year) * 12 +
    (cohorts['order_month'].dt.month - cohorts['first_order_month'].dt.month)
)

# 10. Sort the cohorts by first_order_month and cohort_lifetime
cohorts = cohorts.sort_values(['first_order_month', 'cohort_lifetime'])

# 11. Group the costs data by month and merge it with the cohorts
cost_per_month = costs.groupby('costs_month').agg(costs=('costs', 'sum')).reset_index()

# Merge cost data to the cohorts based on the order_month
cohorts = pd.merge(cohorts, cost_per_month, left_on='order_month', right_on='costs_month')

# 12. Calculate CAC, LTV, and ROMI
cohorts['cac'] = cohorts['costs'] / cohorts['n_buyers']
cohorts['ltv'] = cohorts['revenue'] / cohorts['n_buyers']
cohorts['romi'] = cohorts['ltv'] / cohorts['cac']

# 13. Create a pivot table for ROMI over the cohort lifetime
romi_pivot = cohorts.pivot_table(index='first_order_month', columns='cohort_lifetime', values='romi', aggfunc='mean')

# 14. Calculate the cumulative sum of ROMI for each cohort
cum_sum_romi = romi_pivot.cumsum(axis=1).round(2)

# 15. Plot the ROMI heatmap
sns.set(style='white')
plt.figure(figsize=(13, 9))
plt.title('Cohort de ROMI')
sns.heatmap(cum_sum_romi, annot=True, linewidths=1, linecolor='gray', cmap='coolwarm')
plt.show()

# (opcional) readicionando código anterior para manter códigos posteriores funcionais
orders['order_month'] = orders['buy_ts'].dt.to_period('M')
orders['first_order_month'] = orders.groupby('uid')['order_month'].transform('min')
# 8. Create cohorts based on the first_order_month and calculate revenue and number of buyers
cohorts = orders.groupby(['first_order_month', 'order_month']).agg(
    revenue=('revenue', 'sum'),
    n_buyers=('uid', 'nunique')  # Unique buyers
).reset_index()

margin_rate = 0.4

cohorts['gp'] = cohorts['revenue'] * margin_rate
cohorts['age'] = (cohorts['order_month'] - cohorts['first_order_month'])#.apply(attrgetter('n'))
cohorts['ltv'] = cohorts['gp'] / cohorts['n_buyers']
````
